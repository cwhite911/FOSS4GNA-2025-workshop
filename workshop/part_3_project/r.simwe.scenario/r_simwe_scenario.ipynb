# r.simwe.scenario

This notebook runs the example from r.simwe.scenario's manual page and visualizes the output.

## Setup

We will be using the NC SPM sample location.

```python
import subprocess
import sys

# Ask GRASS where its Python packages are.
sys.path.append(
    subprocess.check_output(["grass", "--config", "python_path"], text=True).strip()
)

# Import GRASS packages
import grass.script as gs
import grass.jupyter as gj
from grass.tools import Tools

# Start GRASS Session
session = gj.init("~/data/grassdata/nc_basic_spm_grass7/user1")
tools = Tools(session=session)
```

## Create example from the tool's documentation

```python
r_simwe_scenario_input = "elevation"
r_simwe_scenario_output = "output"

with gs.RegionManager(raster=r_simwe_scenario_input):
    tools.r_simwe_scenario(input=r_simwe_scenario_input, output=r_simwe_scenario_output)
```

```python
r_simwe_scenario_map = gj.Map()
r_simwe_scenario_map.d_rast(map=r_simwe_scenario_output)
r_simwe_scenario_map.show()
```
